```python
import math

imax=101
jmax=101

dx=1.0/(imax-1)
```

```python
n_max = 200

gamma_n=zeros((4,n_max))

for n in range(1,n_max):
    gamma_n[0,n]=200/(n*pi)*(-cos(n*pi)+1)
    gamma_n[1,n]=200/(n*pi)*(-cos(n*pi)+1)
    gamma_n[2,n]=-200/(n*pi)+200/(n*pi)*(-cos(n*pi)+1)
    gamma_n[3,n]=200/(n*pi)
    
Psi_analytic=zeros((imax,jmax))
    
for i in range(0,imax):
    for j in range(0,jmax):
        for n in range(1,n_max):
            x=i*dx
            y=j*dx
            Psi_analytic[i,j]+=gamma_n[0,n]*sin(n*pi*y)*(cosh(n*pi*x)-sinh(n*pi*x)/tanh(n*pi))
            Psi_analytic[i,j]+=gamma_n[1,n]*sin(n*pi*x)*(cosh(n*pi*(1.-y))-sinh(n*pi*(1.-y))/tanh(n*pi))
            Psi_analytic[i,j]+=gamma_n[2,n]*sin(n*pi*y)*(cosh(n*pi*(1.-x))-sinh(n*pi*(1.-x))/tanh(n*pi))
            Psi_analytic[i,j]+=gamma_n[3,n]*sin(n*pi*x)*(cosh(n*pi*y)-sinh(n*pi*y)/tanh(n*pi))
        
```

```python
fig3=figure()
imshow(Psi_analytic.transpose(),origin='lower')
set_cmap('jet')
colorbar()
```
